```julia
using PyCall
```

```julia
np = pyimport("numpy")
```

```julia
using Plots
using Statistics
using CSV
using RDatasets
using MLBase
using StatsBase
using GLM
```

```julia
df = CSV.read("AAPL Historical Data.csv",DataFrame)
```

```julia
a = np.array(df[!,"Open"])
```

```julia
b = np.array(df[!,"High"])
```

```julia
@time myfit = np.polyfit(a,b,1);
ynew = collect(a) .* myfit[1] .+ myfit[2];

scatter(a,b, color = 2, label = "scatter plot of Open and High Stocks", markerwidth = 0.2)
plot!(a,ynew, color = 0)
```
